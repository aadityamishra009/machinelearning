```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")

```

```python
# Load your dataset (Replace 'dataset.csv' with your file's path)
data = pd.read_csv('dataset.csv')

# Display the first few rows and a summary of the dataset
print("Dataset Overview:")
display(data.head())
print("\nDataset Info:")
data.info()

# Check for missing values
print("\nMissing Values:")
missing_summary = data.isnull().sum().to_frame('Missing Values')
missing_summary['Percentage'] = (missing_summary['Missing Values'] / len(data)) * 100
display(missing_summary)
```

```python
# Statistical description of numerical columns
print("\nStatistical Summary:")
display(data.describe())
```

```python
# Plot missing values
plt.figure(figsize=(10, 6))
missing_summary['Missing Values'].plot(kind='bar', color='skyblue')
plt.title('Missing Values by Column')
plt.ylabel('Number of Missing Values')
plt.xlabel('Columns')
plt.xticks(rotation=45)
plt.show()
```

```python
# Define numerical columns
numerical_columns = [
    'Total Police Expenditure Of Current Year (UOM:INR(IndianRupees))',
    'Total Police Expenditure Increase Or Decrease Over Previous Year (UOM:INR(IndianRupees))',
    'Total Police Training Expenditure Of Current Year (UOM:INR(IndianRupees))',
    'Percentage Of Total Police Training Expenditure To Total Police Expenditure (%) (UOM:%(Percentage))'
]

# Plot histograms for numerical columns
for column in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(data[column].dropna(), kde=True, bins=30, color='blue')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

```

```python
# Compute the correlation matrix
correlation_matrix = data.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Correlation Matrix for Numerical Features')
plt.show()

```

```python
# Scatter plot for Total Police Expenditure vs Training Expenditure
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='Total Police Expenditure Of Current Year (UOM:INR(IndianRupees))',
    y='Total Police Training Expenditure Of Current Year (UOM:INR(IndianRupees))',
    data=data,
    hue='Figures Type'
)
plt.title('Total Expenditure vs. Training Expenditure')
plt.xlabel('Total Police Expenditure (INR)')
plt.ylabel('Training Expenditure (INR)')
plt.legend(title='Figures Type')
plt.show()
```

```python
# Plot state-wise data distribution
plt.figure(figsize=(12, 6))
sns.countplot(y='State', data=data, order=data['State'].value_counts().index, palette="viridis")
plt.title('State-wise Data Distribution')
plt.xlabel('Count')
plt.ylabel('State')
plt.show()
```

```python
# Convert Year to string for grouping
data['Year'] = data['Year'].astype(str)

# Group data by year to calculate mean police expenditure
yearly_expenditure = data.groupby('Year')['Total Police Expenditure Of Current Year (UOM:INR(IndianRupees))'].mean()

# Plot yearly trend
plt.figure(figsize=(12, 6))
yearly_expenditure.plot(kind='line', marker='o', color='teal')
plt.title('Yearly Trend of Mean Police Expenditure')
plt.xlabel('Year')
plt.ylabel('Mean Police Expenditure (INR)')
plt.grid(True)
plt.show()

```
